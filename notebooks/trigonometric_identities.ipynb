# Background

Trigonometric identities are critical for Calculus, Differential Equations, Real and Complex Analysis, Discrete Math, Linear and Nonlinear Statistical Modeling, etc. The problem is these identities are traditionally taught in a memorization sense: e.g. the teacher might derive it on the board once but then says something along the lines of "memorize that $c^2 = a^2 + b^2 - 2ab\cos{\theta}$, it'll be on the test." Students thus memorize it for the upcoming test, then forget it because many students go on to take algebra 2, precalculus, etc. and don't think about it again (even though the underpinning idea is useful) until they get to calculus and finding integrals or derivatives of trigonometric functions more complicated than $\sin$ or $\cos$. This notebook attempts to provide quick to understand visual proofs of the various trigonometric identities.

## Pythagorean Theorem

Most students know the Pythagorean Theorem by heart. It is perhaps one of the most straightforward and conceptually standard building blocks of geometry. It is also a building block of trigonometry and used in the derivation of many various trigonometric identities. That being said, for completeness, we shall show how the Pythagorean Theorem works. The formula is well known: $a^2 + b^2 = c^2$. Below is a demonstration of the rearrangement geometric proof. It shows how by rearrangement of four identical triangles we can visualize the area of an inscribed $c^2$ square is equal to the sum of two squares $a^2$ and $b^2$.

```python
import numpy as np

class PythagoreanTheorem(Scene):
    def construct(self):
        # Leg lengths for a generic right triangle.
        a = 3.2
        b = 2.1
        c = np.hypot(a, b)
        s = a + b

        # Place the (a+b)^2 construction near the center.
        bl = np.array([-s / 2, -s / 2, 0.0])
        br = np.array([s / 2, -s / 2, 0.0])
        tr = np.array([s / 2, s / 2, 0.0])
        tl = np.array([-s / 2, s / 2, 0.0])

        # Initial right triangle at bottom-left corner of the future large square.
        p0 = bl
        p1 = bl + np.array([b, 0.0, 0.0])
        p2 = bl + np.array([0.0, a, 0.0])

        tri_bl = Polygon(p0, p1, p2, color=BLUE, fill_opacity=0.35)

        # Side labels a, b, c on the initial triangle.
        a_label = MathTex("a").scale(0.8).next_to(Line(p0, p2), LEFT, buff=0.15)
        b_label = MathTex("b").scale(0.8).next_to(Line(p0, p1), DOWN, buff=0.15)
        c_label = MathTex("c").scale(0.8).move_to((p1 + p2) / 2 + 0.28 * normalize(np.array([a, b, 0.0])))

        self.play(Create(tri_bl), FadeIn(a_label, b_label, c_label))
        self.wait(1)

        # Build the square on the hypotenuse, outward from the triangle's right angle.
        hyp = p2 - p1
        hyp_unit = hyp / np.linalg.norm(hyp)
        n1 = np.array([hyp_unit[1], -hyp_unit[0], 0.0])
        n2 = -n1

        # Choose the normal that points away from the right-angle vertex p0.
        if np.linalg.norm((p1 + c * n1) - p0) > np.linalg.norm((p1 + c * n2) - p0):
            n = n1
        else:
            n = n2

        q1 = p1
        q2 = p2
        q3 = p2 + c * n
        q4 = p1 + c * n

        c_square = Polygon(q1, q2, q3, q4, color=YELLOW, fill_opacity=0.25)
        c2_label = MathTex("c^2").scale(0.9).move_to(c_square.get_center())

        self.play(Create(c_square), FadeIn(c2_label))

        # Add the other three congruent triangles to surround the c^2 square.
        tri_br = Polygon(br, br + np.array([0.0, b, 0.0]), br + np.array([-a, 0.0, 0.0]), color=BLUE, fill_opacity=0.35)
        tri_tr = Polygon(tr, tr + np.array([-b, 0.0, 0.0]), tr + np.array([0.0, -a, 0.0]), color=BLUE, fill_opacity=0.35)
        tri_tl = Polygon(tl, tl + np.array([0.0, -b, 0.0]), tl + np.array([a, 0.0, 0.0]), color=BLUE, fill_opacity=0.35)
        self.play(FadeIn(tri_br, tri_tr, tri_tl))

        outer_square = Polygon(bl, br, tr, tl, color=WHITE)

        self.play(Create(outer_square))

        self.wait(0.5)

        # First pair target: bottom-left a*b rectangle.
        r1_bl = bl
        r1_br = bl + np.array([a, 0.0, 0.0])
        r1_tl = bl + np.array([0.0, b, 0.0])
        r1_tr = bl + np.array([a, b, 0.0])

        tri_bl_target = Polygon(r1_bl, r1_br, r1_tl, color=BLUE, fill_opacity=0.35)
        tri_tr_target = Polygon(r1_tr, r1_br, r1_tl, color=BLUE, fill_opacity=0.35)
        rect1_outline = Rectangle(width=a, height=b, color=GREEN).move_to((r1_bl + r1_tr) / 2)

        self.play(
            Transform(tri_bl, tri_bl_target),
            Transform(tri_tr, tri_tr_target),
            FadeOut(c_square, c2_label, a_label, b_label, c_label),
            run_time=1.8,
        )
        self.play(Create(rect1_outline), run_time=0.9)

        # Second pair target: top-right a*b rectangle.
        r2_bl = bl + np.array([a, b, 0.0])
        r2_br = bl + np.array([s, b, 0.0])
        r2_tl = bl + np.array([a, s, 0.0])
        r2_tr = tr

        tri_br_target = Polygon(r2_bl, r2_br, r2_tl, color=BLUE, fill_opacity=0.35)
        tri_tl_target = Polygon(r2_tr, r2_br, r2_tl, color=BLUE, fill_opacity=0.35)
        rect2_outline = Rectangle(width=b, height=a, color=GREEN).move_to((r2_bl + r2_tr) / 2)

        self.play(
            Transform(tri_br, tri_br_target),
            Transform(tri_tl, tri_tl_target),
            run_time=1.8,
        )
        self.play(Create(rect2_outline), run_time=0.9)

        # Reveal the two remaining squares a^2 and b^2.
        a_sq = Square(side_length=a, color=TEAL, fill_opacity=0.2)
        a_sq.move_to((bl + np.array([a / 2, b + a / 2, 0.0])))
        b_sq = Square(side_length=b, color=ORANGE, fill_opacity=0.2)
        b_sq.move_to((bl + np.array([a + b / 2, b / 2, 0.0])))

        a2_label = MathTex("a^2").scale(0.85).move_to(a_sq.get_center())
        b2_label = MathTex("b^2").scale(0.85).move_to(b_sq.get_center())

        self.play(
            Create(a_sq),
            Create(b_sq),
            FadeIn(a2_label, b2_label),
            run_time=1.2,
        )

        self.wait(1)

# Render inline-friendly quality in notebook context.
config.quality = "medium_quality"
config.preview = False

scene = PythagoreanTheorem()
scene.render()

from IPython.display import Video
Video(scene.renderer.file_writer.movie_file_path, embed=True)
```

# Sine, Cosine, and Tangent

Following the Pythagorean Theorem, perhaps the second most recognizable part of trig is the "trigonometric functions". The three core functions are $\sin$, $\cos$, and $\tan$. They are mathematically defined functions that are commonly visualized in one of two methods:
1. The relationship to an angle (usually defined $\theta$) within a right triangle
2. A representation of a Euclidean coordinate on the unit circle

## Right Triangle

Because most students taking trig are familiar with triangles and the Euclidean coordinate system (e.g. a grid), right triangles are the most traditional introduction to the trig functions. They are defined as follows on the triangle below:

# Identities

Now we have all the necessary information to prove the trig identities (they're literally all just relationships that bounce the understanding of trig functions between unit circles and right triangles, then applying some version of the pythagorean theorem).

## Pythagorean Identity

Perhaps the most common trig identity is the Pythagorean Identity. It is: $\sin^2{\theta} + \cos^2{\theta} = 1$. This can be proven any number of ways. We shall choose two very quick proofs.

### Proof Using Right Angle Triangles

First, we'll prove the identity via right triangles.

1. Square the definition of sine: $\sin^2{\theta} = \frac{\text{opposite}^2}{\text{hypotenuse}^2}$
2. Square the definition of cosine: $\cos^2{\theta} = \frac{\text{adjacent}^2}{\text{hypotenuse}^2}$
3. Add the right hand sides of the equations together: $\sin^2{\theta} + \cos^2{\theta} = \frac{\text{opposite}^2 + \text{adjacent}^2}{\text{hypotenuse}^2}$
4. Put that aside and look at the Pythagorean Theorem. Divide both sides by $c^2$: $1 = \frac{a^2 + b^2}{c^2}$
5. Realize 3 and 4 are the same equation: $\frac{\text{opposite}^2 + \text{adjacent}^2}{\text{hypotenuse}^2} = \frac{b^2 + a^2}{c^2}$
6. Take the left hand sides of the equations defined in 3 and 4 to get the Pythagorean Identity: $\sin^2{\theta} + \cos^2{\theta} = 1$

### Proof Using the Unit Circle

Another example of the unit circle being slightly easier to utilize:

1. The definition of a circle on a Euclidean Plane is $x^2 + y^2 = 1$
2. Via the definition of a unit circle, we can substitute $x = \cos$ and $y = \sin$
3. The resulting substitution gives us the identity: $\sin^2{\theta} + \cos^2{\theta} = 1$

## Law of Cosines

The first identity we shall derive will be the Law of Cosines. It states that $c^2 = a^2 + b^2 - 2ab\cos{C}$

```python
class LawOfCosines(Scene):
    def construct(self):
        # Large scalene triangle with a horizontal base (side c).
        pA = np.array([-5.0, -2.0, 0.0])
        pB = np.array([5.0, -2.0, 0.0])
        pC = np.array([1.6, 2.5, 0.0])

        triangle = Polygon(pA, pB, pC, color=WHITE, stroke_width=4)
        self.play(Create(triangle))

        side_a = Line(pB, pC)  # Opposite angle A
        side_b = Line(pA, pC)  # Opposite angle B
        side_c = Line(pA, pB)  # Opposite angle C

        label_a = MathTex("a").next_to(side_a, RIGHT, buff=0.01)
        label_b = MathTex("b").next_to(side_b, LEFT, buff=0.05)
        label_c = MathTex("c").next_to(side_c, DOWN, buff=0.05)

        angle_A = MathTex("A").next_to(pA, DOWN + LEFT, buff=0.1)
        angle_B = MathTex("B").next_to(pB, DOWN + RIGHT, buff=0.1)
        angle_C = MathTex("C").next_to(pC, UP, buff=0.1)

        self.play(
            FadeIn(label_a),
            FadeIn(label_b),
            FadeIn(label_c),
            FadeIn(angle_A),
            FadeIn(angle_B),
            FadeIn(angle_C),
        )

        # Altitude from C to side c (AB), labeled h.
        pD = np.array([pC[0], pA[1], 0.0])
        altitude = DashedLine(pC, pD, dash_length=0.18, color=YELLOW)
        h_label = MathTex("h").next_to(altitude.get_center(), RIGHT, buff=0.15)

        # Right-angle marker at the foot of the altitude.
        marker_size = 0.32
        marker = Polygon(
            pD,
            pD + np.array([marker_size, 0.0, 0.0]),
            pD + np.array([marker_size, marker_size, 0.0]),
            pD + np.array([0.0, marker_size, 0.0]),
            color=YELLOW,
            fill_opacity=0,
            stroke_width=2,
        )

        self.play(Create(altitude), FadeIn(h_label), FadeIn(marker))

        # Split the base c into x (right segment) and c - x (left segment).
        right_base = Line(pD, pB)
        left_base = Line(pA, pD)
        label_x = MathTex("x").next_to(right_base, DOWN, buff=0.25)
        label_c_minus_x = MathTex("c - x").next_to(left_base, DOWN, buff=0.25)

        # Fill the two right triangles created by the altitude with distinct colors.
        left_triangle_fill = Polygon(
            pA, pD, pC, color=BLUE, fill_opacity=0.25, stroke_opacity=0
        )
        right_triangle_fill = Polygon(
            pD, pB, pC, color=GREEN, fill_opacity=0.25, stroke_opacity=0
        )

        self.play(
            FadeOut(label_c),
            FadeIn(label_x),
            FadeIn(label_c_minus_x),
            FadeIn(left_triangle_fill),
            FadeIn(right_triangle_fill),
        )

        self.wait(1.5)
        triangle_group = VGroup(
            triangle,
            label_a,
            label_b,
            angle_A,
            angle_B,
            angle_C,
            altitude,
            h_label,
            marker,
            label_x,
            label_c_minus_x,
            left_triangle_fill,
            right_triangle_fill,
        )
        self.play(triangle_group.animate.scale(0.5).to_edge(LEFT).shift(DOWN * 0.15))

        # Keep the algebra compact on the right side and transform one equation at a time.
        eq_rhs_note = MathTex("{{h}}^2", "=", "{{a}}^2", "-", "{{x}}^2").to_edge(RIGHT)
        eq_main = MathTex(
            "{{h}}^2", "+", "({{c}} - {{x}})^2", "=", "{{b}}^2"
        ).next_to(eq_rhs_note, DOWN).to_edge(RIGHT)
        eq_main.set_color(BLUE)
        eq_rhs_note.set_color(GREEN)

        self.play(Write(eq_rhs_note), Write(eq_main))

        self.wait(0.5)
        eq_substituted = MathTex(
            "{{a}}^2", "-", "{{x}}^2", "+", "({{c}} - {{x}})^2", "=", "{{b}}^2"
        ).next_to(eq_main, DOWN).to_edge(RIGHT)
        self.play(Write(eq_substituted))

        self.wait(0.5)
        self.play(FadeOut(eq_main), FadeOut(eq_rhs_note), run_time=0.8)

        self.wait(0.5)
        eq_expanded = MathTex(
            "{{a}}^2", "-", "{{x}}^2", "+", "{{c}}^2", "-", "2{{c}}{{x}}", "+", "{{x}}^2", "=", "{{b}}^2"
        ).to_edge(RIGHT)
        self.play(Write(eq_expanded), FadeOut(eq_substituted))

        self.wait(0.5)
        eq_reduced = MathTex(
            "{{a}}^2", "+", "{{c}}^2", "-", "2{{c}}{{x}}", "=", "{{b}}^2"
        ).to_edge(RIGHT)
        eq_x_relation = MathTex(r"x = b\cos(\theta)").next_to(eq_reduced, DOWN)
        self.play(Write(eq_reduced), FadeOut(eq_expanded), Write(eq_x_relation))
        
        self.wait(0.5)
        eq_final = MathTex(r"c^2 = a^2 + b^2 - 2ab\cos(\theta)").to_edge(RIGHT)
        self.play(FadeOut(eq_x_relation), Transform(eq_reduced, eq_final))

        self.wait(1.5)

# Render inline-friendly quality in notebook context.
config.quality = "medium_quality"
config.preview = False

scene = LawOfCosines()
scene.render()

from IPython.display import Video
Video(scene.renderer.file_writer.movie_file_path, embed=True)
```
