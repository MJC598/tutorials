## Example animation cell

Run the next cell to preview a simple Manim scene in the notebook.

```python
%%manim -qm IntroCircle

class IntroCircle(Scene):
    def construct(self):
        circle = Circle(color=BLUE)
        label = Text("Core Idea").scale(0.5).next_to(circle, DOWN)

        self.play(Create(circle))
        self.play(FadeIn(label))
        self.wait(1)
```
